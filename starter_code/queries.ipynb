```python
# Las respuestas las voy a mostrar en dataframes, ya que se visualiza mucho mejor. Para pasar los cursores a dataframe,
# voy a usar esta función:

def cursor_a_dataframe(cursor):
    display(pd.DataFrame(list(cursor)))
```

# Answers
### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
babelgum = db.companies.find({"name": "Babelgum"}, {"_id": 0, "name": 1})
```

```python
list(babelgum)
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
many_employees = db.companies.find({"number_of_employees": {"$gt": 5000}}, 
                                   {"_id" : 0, "name": 1, "number_of_employees": 1}
                                  ).sort("number_of_employees", pymongo.ASCENDING
                                        ).limit(15)
```

```python
print(cursor_a_dataframe(many_employees))
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
early_century = db.companies.find({"founded_year": {"$gte": 2000, "$lte": 2005}}, 
                                  {"_id": 0, "name": 1, "founded_year": 1}
                                 ).limit(15)
```

```python
print(cursor_a_dataframe(early_century))
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
valuable_new = db.companies.find({"$and": 
                                  [{"ipo.valuation_amount": {"$gt": 100000000}}, 
                                   {"founded_year": {"$lt": 2010}}]}, 
                                 {"_id": 0, "name": 1, "ipo.valuation_amount": 1}
                                ).sort("ipo.valuation_amount", pymongo.ASCENDING
                                      ).limit(15)
```

```python
print(cursor_a_dataframe(valuable_new))
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
print(cursor_a_dataframe(small_old))
```

### 12. All the companies that have been 'deadpooled' after the third year.

```python
late_deadpool = db.companies.aggregate([{"$project": 
                                       {"_id": 0, "name": 1, "founded_year": 1, "deadpooled_year": 1, 
                                       "difference": {"$subtract": ["$deadpooled_year", "$founded_year"]}}},
                                      {"$match": {"difference": {"$gt": 3}}}, 
                                       {"$limit": 15 }])
```

```python
print(cursor_a_dataframe(late_deadpool))
```

### 20. All the companies that have been founded between 2000 and 2010, but have not been acquired before 2011.

```python
earlycentury_stillalive = db.companies.find({"$and":
                                             [{"founded_year": {"$gt": 2000}},
                                              {"founded_year": {"$lt": 2010}},
                                              {"$or":
                                              [{"acquisition.acquired_year": {"$gte": 2011}}, 
                                               {"acquisition.acquired_year": None}]}]}, 
                                            {"_id": 0, "name": 1, "founded_year": 1, "acquisition.acquired_year": 1}
                                           ).limit(15)
```

```python
print(cursor_a_dataframe(earlycentury_stillalive))
```
